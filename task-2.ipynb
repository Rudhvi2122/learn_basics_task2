```python
#importing pandas
import pandas as pd

#Reading The input
data_set = pd.read_excel(r"C:\Users\Dell\Downloads\Input_1 - Python Developer Intern - Task 2 - Datasets.xlsx")
data_set
```

```python
test_names = ["Concept Test 1", "Concept Test 2", "Concept Test 3", "Concept Test 4", "Concept Test 5", "Full Chapter Test 1", "Full Chapter Test 2", "Topic Test 1", "Topic Test 2"]

#Creaing New Dataframe according to the output
empty_dataframe = pd.DataFrame()
empty_dataframe["Name"] = data_set["Name"]
empty_dataframe["username"] = data_set["id"]
empty_dataframe["Chapter Tag"] = data_set["Chapter Tag"]

#creating empty list for concat
data_set_list = []
for i in range(len(test_names)):
    test_data = empty_dataframe.copy()
    test_data["test_name"] = test_names[i]
    test_data["answered"] = data_set[test_names[i]+" - answered"]
    test_data["correct"] = data_set[test_names[i]+"- correct"]
    test_data["score"] = data_set[test_names[i]+" - score"]
    test_data["skipped"] = data_set[test_names[i]+"- skipped"]
    test_data["time-taken (seconds)"] = data_set[test_names[i]+" - time-taken (seconds)"]
    test_data["wrong"] = data_set[test_names[i]+"- wrong"]
    #storing all tests data into a list
    data_set_list.append(test_data)

#Combining all the tests data into one
new_data_set = pd.concat(data_set_list, ignore_index=True)
new_data_set
```

```python
#Removing null values
new_data_set = new_data_set[new_data_set["answered"] != "-"]

#Sorting data based on username and test name
new_data_set = new_data_set.sort_values(by=["username", "test_name"], ignore_index=True)
new_data_set
```
